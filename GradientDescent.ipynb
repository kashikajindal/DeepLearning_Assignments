```python
import numpy as np
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
columns = ["Sex","Length","Diameter","Height","Whole_weight",
           "Shucked_weight","Viscera_weight","Shell_weight","Rings"]

df = pd.read_csv(url, header=None, names=columns)

print("\n========== DATASET INFO ==========")
print(f"Total Rows: {len(df)}")
print("\nColumns:")
for col in df.columns:
    print(f" - {col}")
print("\nFirst 5 Rows:")
print(df.head())

y = df["Rings"].values.reshape(-1,1) + 1.5
X = df[["Length","Diameter","Shell_weight"]].values

N = X.shape[0]
split = int(0.8 * N)

X_train = X[:split]
X_test = X[split:]
y_train = y[:split]
y_test = y[split:]

print("\n========== TRAIN-TEST SPLIT ==========")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape : {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape : {y_test.shape}")

mean = X_train.mean(axis=0)
std = X_train.std(axis=0)

X_train = (X_train - mean) / std
X_test = (X_test - mean) / std

def forward(X, w, b):
    y_hat = np.dot(X, w) + b
    return y_hat

def mse(y, y_hat):
    return np.mean((y - y_hat) ** 2)

def mae(y, y_hat):
    return np.mean(np.abs(y - y_hat))

def grad_w(X, y, y_hat):
    N = X.shape[0]
    return (-2/N) * np.dot(X.T, (y - y_hat))

def grad_b(y, y_hat):
    N = y.shape[0]
    return (-2/N) * np.sum(y - y_hat)

d = X_train.shape[1]
w = np.random.randn(d,1) * 0.01
b = 0.0

lr = 0.01
epochs = 200

print("\n========== TRAINING STARTED ==========")

for epoch in range(epochs):
    y_hat = forward(X_train, w, b)
    loss = mse(y_train, y_hat)
    dW = grad_w(X_train, y_train, y_hat)
    db = grad_b(y_train, y_hat)
    w = w - lr * dW
    b = b - lr * db
    if epoch % 20 == 0:
        print(f"Epoch {epoch:03d} | Loss: {loss:.6f}")

print("\n========== TRAINING COMPLETED ==========")

y_test_hat = forward(X_test, w, b)

test_mse = mse(y_test, y_test_hat)
test_mae = mae(y_test, y_test_hat)

print("\n========== EVALUATION ==========")
print(f"Test MSE : {test_mse:.6f}")
print(f"Test MAE : {test_mae:.6f}")

print("\n========== SAMPLE PREDICTIONS ==========")
for i in range(5):
    true_val = y_test[i][0]
    pred_val = y_test_hat[i][0]
    error = abs(true_val - pred_val)
    print(f"Sample {i+1}:")
    print(f"  True Age      : {true_val:.3f}")
    print(f"  Predicted Age : {pred_val:.3f}")
    print(f"  Absolute Error: {error:.3f}\n")
```
